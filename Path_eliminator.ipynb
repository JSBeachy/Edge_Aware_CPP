```python
import numpy as np
from scipy.spatial import KDTree
import open3d as o3d
from PCAClass import PCABounding, Best_Fit_CPP


def pseudo_binary_order(n):
    """Generate pseudo-binary ordering for n passes."""
    if n == 0:
        return []
    order = [0, n-1]
    queue = [(0, n-1)]
    while queue:
        start, end = queue.pop(0)
        mid = (start + end) // 2
        if mid not in order:
            order.append(mid)
        if start + 1 < mid:
            queue.append((start, mid))
        if mid + 1 < end:
            queue.append((mid, end))
    
    return [i for i in order if i < n]

def check_redundancy(passes, coverage_radius):
    ordered_indices = pseudo_binary_order(len(passes))
    pass_kdtrees = []  # List to store KD-trees for each analyzed pass
    redundancy_masks = []
    
    for idx in ordered_indices:
        pass_points = passes[idx]
        redundant_mask = np.zeros(pass_points.shape[0], dtype=bool)

        # Check against all previous passes
        count_covered = np.zeros(pass_points.shape[0], dtype=int)
        for tree in pass_kdtrees:
            dists, _ = tree.query(pass_points, distance_upper_bound=coverage_radius)
            count_covered += dists < coverage_radius  # Increment count if covered by a pass
        
        redundant_mask = count_covered >= 2  # Require coverage from at least two passes

        # Store the new pass as a KD-tree
        if pass_points.size > 0:
            pass_kdtrees.append(KDTree(pass_points)) 
        
        redundancy_masks.append(redundant_mask)
    
    return redundancy_masks

# Example Usage:
#passes = [np.random.rand(100, 3) * 10 for _ in range(8)]  # Simulated passes
#coverage_radius = 0.5
#redundancy_masks = check_redundancy(passes, coverage_radius)

# Print how many redundant points were found in each pass
#for i, mask in enumerate(redundancy_masks):
    print(f"Pass {i}: {np.sum(mask)} redundant points")

segment=Best_Fit_CPP("plane_segments\Hyper_meshed_noise.stl")
path1=np.linspace([75,0,0],[75,400,0],51)
path2=np.linspace([100,0,0],[125,400,0],51)
path3=np.linspace([175,0,0],[175,400,0],51)
colors1 = np.full((51, 3), (0, 1, 0), dtype=np.float32)  
colors2 = np.full((51, 3), (0, 0.5, 0.5))  
colors3 = np.full((51, 3), (0, 0, 1)) 

lines=np.array([path1,path2,path3])
colors=[colors1,colors2,colors3]
adjusted_lines = np.array([line + np.array([0, 0, 10]) for line in lines])
print(check_redundancy(adjusted_lines, coverage_radius=32))

trial=o3d.geometry.PointCloud()
trial.points=o3d.utility.Vector3dVector(np.vstack(adjusted_lines))
trial.colors=o3d.utility.Vector3dVector(np.vstack(colors))
o3d.visualization.draw_geometries([segment.mesh,segment.bounding_box, trial,],mesh_show_back_face=True)
```
